```python
import wikipedia
import re
import wordcloud
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt

```

```python
#importing page from wikipedia

wiki=wikipedia.page('Roman Empire')
text=wiki.content

text=re.sub(r'==.*?==+','',text)
text=text.replace('\n','')

```

```python
def plt_cloud(wordcloud):
    plt.figure(figsize=(60,50))
    plt.imshow(wordcloud)
    plt.axis("off")
    
```

```python
# Display your wordcloud image
from wordcloud import WordCloud,STOPWORDS

mask=np.array(Image.open(r'C:\Users\sauna\OneDrive\Pictures\crown.png'))

wordcloud=WordCloud(width=700, height=500, random_state=1, background_color='red',colormap='YlOrBr',collocations=False,stopwords=STOPWORDS,mask=mask).generate(text)

#plot
plt_cloud(wordcloud)
```
